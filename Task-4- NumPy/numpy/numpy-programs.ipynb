# NumPy Practice Notebook

This notebook demonstrates basic NumPy operations including:
- Array creation
- Indexing and slicing
- Mathematical operations
- Reshaping
- Broadcasting
- Array manipulation
- Random number generation

```python
import numpy as np
```

## Creating a 1D Array


```python
arr1 = np.array([10, 20, 30, 40, 50])
print(arr1)
```

## Creating a 2D Array
# contains 2 rows and 3 columns

```python
arr2 = np.array([[1, 2, 3],
                 [4, 5, 6]])
print(arr2)
```

## Creating a 3D Array
# contains 2 rows and 2 columns

```python
arr3 = np.array([
    [[1, 2], [3, 4]],
    [[5, 6], [7, 8]]
])
print(arr3)
```

## Array Dimensions
## dimension refer to as number of indexes used to acces an element 

```python
print(arr1.ndim)
print(arr2.ndim)
print(arr3.ndim)
```

```python
## Array Shape
## shape refer to as rows and columns present in array
print(arr1.shape)
print(arr2.shape)
print(arr3.shape)
```

```python
## Creating Arrays Using arange()
## arange() functions is used to create array 

arr = np.arange(1, 11)
print(arr)
```

```python
## Array Indexing
## indexing is refer to as accessing elements using indexes from array
arr = np.array([10, 20, 30, 40, 50])
print(arr[0])
print(arr[3])
```

```python
## Negative Indexing
## negative indexing start from end point of array i.e last element 

print(arr[-1])
print(arr[-2])
```

```python
## Array Slicing
## slicing refers to as using index i.e start and end index(end index is not included) to access part of array
print(arr[1:4])
```

```python
## 2D Array Indexing
## accessing elemnt from 1 row and second column
matrix = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

print(matrix[1, 2])
```

```python
## operation on arrays
## Addition of Arrays
## add elemnts column wise
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

print(a + b)
```

```python
## Subtraction of Arrays
## subtracts elements from b to a
print(a - b)
```

```python
## Multiplication of Arrays
print(a * b)
```

```python
## Division of Arrays
print(a / b)
```

```python
## Sum of Array Elements
## adds elements of an array
arr = np.array([1, 2, 3, 4, 5])
print(np.sum(arr))
```

```python
## Mean of Array Elements
print(np.mean(arr))
```

```python
## Standard Deviation

print(np.std(arr))
```

```python
## Reshaping Arrays
## reshaping is referred to as changing 1D array to another Dimensonal array like 2D and 3D
arr = np.arange(12)
reshaped = arr.reshape(3, 4)

print(reshaped)
```

```python
## Flattening Arrays
## flattening of arrays means to regain array in original shape 
flat = reshaped.flatten()
print(flat)
```

```python
## Broadcasting Example
## Adding a scalar value to every elemen i.e it adds 1 to every element of the array
matrix = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

print(matrix + 10)
```

```python
## Broadcasting is the process by which NumPy automatically expands smaller arrays to match the shape of larger arrays so that arithmetic operations can be performed element-wise.
A = np.array([[1, 2, 3],
              [4, 5, 6]])

B = np.array([10, 20, 30])

print(A + B)
```

```python
## Broadcasting Between Arrays
## here array a is 2D array with shape (3,1) that is 3 rows and 1 column and b copies it for each row and add elemts from array a
a = np.array([[1], [2], [3]])
b = np.array([10, 20, 30])

print(a + b)
```

```python
## Array Concatenation
## concatenation of arrays can be defined as adding one array with other
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

combined = np.concatenate((a, b))
print(combined)
```

```python
## Finding Maximum and Minimum Values
arr = np.array([15, 7, 22, 3, 18])

print(np.max(arr))
print(np.min(arr))
```

```python
## Generating Random Numbers
## np.random.randint prints any random number with syntax (low,high,size)
random_ints = np.random.randint(1, 100, 10)
print(random_ints)
```
