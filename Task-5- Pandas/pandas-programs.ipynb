## Creating a DataFrame
## A DataFrame is a two-dimensional table with rows and columns.
## Created Employee Table For Performing  Operations

```python
import pandas as pd
import numpy as np

data = {
    "Name": [
        "Ali", "Sara", "John", "David", "Emma",
        "Aisha", "Rahul", "Priya", "Ahmed", "Neha",
        "Arjun", "Fatima", "Karan", "Zoya", np.nan
    ],

    "Age": [
        21, 22, 23, np.nan, 25,
        24, 26, 27, np.nan, 29,
        30, 28, 31, np.nan, 32
    ],

    "Department": [
        "IT", "HR", "IT", "Finance", "HR",
        "Marketing", "IT", np.nan, "Finance", "HR",
        "Marketing", "IT", "Finance", "HR", "IT"
    ],

    "Salary": [
        30000, 35000, 40000, 45000, np.nan,
        38000, 42000, 39000, 47000, np.nan,
        52000, 43000, 48000, 51000, 55000
    ]
}

df = pd.DataFrame(data)

print(df)
```

## Reading a CSV File
## Use read_csv() to load CSV data. 

```python
# Example
# df = pd.read_csv("employees.csv")

print(df)
```

## Display First 5 Rows
## using head() to print first 5 rows from data

```python
print(df.head())
```

## Display Last 5 Rows
## tail() shows the last 5 records.

```python
print(df.tail())
```

## Data Information
## info() displays column details and data types. and also indexes 

```python
print(df.info())
```

## Statistical Summary
## describe() provides statistical information like count , mean or average minimum value and maximum value 

```python
print(df.describe())
```

## Filtering Rows
## used to select the rows that satisfies a condition
## Display employees with salary greater than 40000.

```python
print(df[df["Salary"] > 40000])
```

## Filtering Multiple Conditions

```python
print(df[(df["Salary"] > 35000) & (df["Department"] == "IT")])
```

## Adding a New Column

```python
df["Bonus"] = df["Salary"] * 0.10

print(df)
```

## Handling Missing Values
## Creating sample missing values.

```python
df.loc[2, "Bonus"] = np.nan

print(df)
```

## Detect Missing Values
## isnull() checks every cell in the DataFrame and asks:
## Is this value missing NaN/null

```python
print(df.isnull())
```

## Fill Missing Values
## fillna(value)  Replaces missing values with a specified value

```python
df["Bonus"] = df["Bonus"].fillna(0)

print(df)
```

## Drop Missing Values

```python
print(df.dropna())
```

## Group By Operation is used to group rows that have same value sin column and perform calculations on each group
Calculate average salary department-wise.

```python
print(df.groupby("Department")["Salary"].mean())
```

## Merging DataFrames is same as merging two datasets 
## on department here means it will  Match rows from both DataFrames using the Department column
## Pandas looks at each department in df and finds the corresponding location in department_info.

```python
department_info = pd.DataFrame({
    "Department": ["IT", "HR", "Finance"],
    "Location": ["Hyderabad", "Mumbai", "Delhi"]
})

merged_df = pd.merge(df, department_info, on="Department")

print(merged_df)
```

## Creating a Pivot Table
## A Pivot Table is used to summarize and analyze data by grouping rows and calculating values such as sum, average, count

```python
pivot = pd.pivot_table(
    df,
    values="Salary",
    index="Department",
    aggfunc="mean"
)

print(pivot)
```
