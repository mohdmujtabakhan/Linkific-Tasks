# Salary Prediction System using Linear Regression
## Objective
The objective of this project is to build a Machine Learning model that predicts employee salaries based on years of experience using the Linear Regression algorithm.

This project demonstrates the complete Machine Learning workflow including:

- Loading a dataset
- Data cleaning and preprocessing
- Handling missing values
- Splitting data into training and testing sets
- Training a Linear Regression model
- Making predictions
- Evaluating model performance
- Saving the trained model using Pickle and Joblib

# Import Required Libraries

The following libraries are used:

- Pandas → Data manipulation
- NumPy → Numerical operations
- Scikit-Learn → Machine Learning algorithms
- Pickle → Save trained models
- Joblib → Efficient model storage

```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

import pickle
import joblib
```

# Load Dataset

The dataset is loaded using Pandas.
The dataset contains employee information such as:
- Age
- Gender
- Education Level
- Job Title
- Years of Experience
- Salary

```python
df = pd.read_csv("Salary_Data.csv")

df.head()
```

# Explore Dataset

Before training a model, it is important to understand:
- Number of rows and columns
- Data types
- Missing values
- Statistical summary

```python
print(df.shape)

df.info()

df.describe()
```

# Handling Missing Values

Missing values can negatively affect model performance.
We first identify missing values and then remove them.

```python
df.isnull().sum()
```

# Data Cleaning

Rows containing missing values are removed to improve data quality.
Data cleaning is an important preprocessing step in Machine Learning.

```python
df = df.dropna()

print("Dataset Shape After Cleaning:")
print(df.shape)
```

# Feature Selection
For simplicity, we use:

Feature:(X)
- Years of Experience

Target:(Y)
- Salary
The model will learn how salary changes with experience.

```python
X = df[['Years of Experience']]

y = df['Salary']
```

# Split Dataset

The dataset is divided into:

- 80% Training Data
- 20% Testing Data

The training set is used to train the model.

The testing set is used to evaluate the model.

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42
)

print("Training Samples:", len(X_train))
print("Testing Samples:", len(X_test))
```

# Linear Regression Model

Linear Regression attempts to find the best-fit line between input features and target values.

Formula:
Salary = m × Experience + c

Where:

- m = Slope         X=features (Years of experience)
- c = Intercept     Y=Target  (Salary)

```python
model = LinearRegression()

model.fit(X_train, y_train)

print("Model Trained Successfully")
```

# Model Prediction

After training, the model can predict salaries for unseen data.

```python
predictions = model.predict(X_test)

predictions[:10]
```

# Predict Salary for New Employee

The trained model can estimate salary based on years of experience.

```python
import pandas as pd

experience = pd.DataFrame({
    'Years of Experience': [7]
})

predicted_salary = model.predict(experience)

print("Predicted Salary:", predicted_salary[0])
```

# Evaluation Metrics

The model is evaluated using:
## Mean Squared Error (MSE)
Measures average squared prediction error.

Lower value is better.

## Root Mean Squared Error (RMSE)
Square root of MSE.
Represents prediction error in actual salary units.

Lower value is better.

## R² Score
Measures how well the model explains the data.

Range:
- 1.0 → Perfect Model
- 0.8+ → Good Model
- 0.5+ → Acceptable Model

```python
mse = mean_squared_error(y_test, predictions)

rmse = np.sqrt(mse)

r2 = r2_score(y_test, predictions)

print("MSE :", mse)
print("RMSE:", rmse)
print("R² Score:", r2)
```

# Save Model using Pickle

Pickle allows us to save the trained model and use it later without retraining.

```python
with open("salary_model.pkl", "wb") as file:
    pickle.dump(model, file)

print("Model Saved Successfully")
```
