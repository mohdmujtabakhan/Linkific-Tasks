# Analysis 1: Who Survived More - Men or Women?

## Objective
Determine whether male or female passengers had a higher number of survivors.

## Why Bar Chart?
A bar chart is used because we are comparing survivor counts across two categories:
- Male
- Female

## Method
1. Filter passengers who survived.
2. Count survivors by gender.
3. Visualize the counts using a bar chart.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("train.csv")

# Count survivors by gender
survival_gender = df[df["Survived"] == 1]["Sex"].value_counts()

# Plot
plt.figure(figsize=(6,4))
plt.bar(survival_gender.index, survival_gender.values)

plt.title("Number of Survivors by Gender")
plt.xlabel("Gender")
plt.ylabel("Number of Survivors")

plt.show()
```

# Analysis 2: Average Age by Passenger Class

## Objective
Calculate the average age of passengers in each travel class.

## Passenger Classes
- Class 1 = First Class
- Class 2 = Second Class
- Class 3 = Third Class

## Why Bar Chart?
A bar chart helps compare average ages across passenger classes.

## Method
1. Group data by passenger class.
2. Calculate average age using mean().
3. Display results using a bar chart.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("train.csv")

# Average age by class
avg_age = df.groupby("Pclass")["Age"].mean()

# Plot
plt.figure(figsize=(6,4))
plt.bar(avg_age.index, avg_age.values)

plt.title("Average Age by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Average Age")

plt.show()
```

# Analysis 3: Survival Rate by Embarkation Port

## Objective
Analyze how survival rates varied based on the passenger's embarkation port.

## Embarkation Ports
- C = Cherbourg
- Q = Queenstown
- S = Southampton

## Why Bar Chart?
A bar chart allows easy comparison of survival percentages across embarkation ports.

## Method
1. Group passengers by embarkation port.
2. Calculate survival percentage.
3. Visualize the results using a bar chart.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("train.csv")

# Calculate survival rate
survival_rate = df.groupby("Embarked")["Survived"].mean() * 100

# Plot
plt.figure(figsize=(6,4))
plt.bar(survival_rate.index, survival_rate.values)

plt.title("Survival Rate by Embarkation Port")
plt.xlabel("Embarkation Port")
plt.ylabel("Survival Rate (%)")

plt.show()
```
