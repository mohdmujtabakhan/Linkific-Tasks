# Titanic Dataset Visualization using Seaborn
## Aim
To perform exploratory data analysis (EDA) on the Titanic dataset using Seaborn visualizations.
## Objectives
- Understand passenger demographics
- Analyze survival patterns
- Explore relationships between variables
- Learn different Seaborn plot types
## Libraries Used
```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("train.csv")

df.head()
````

# Heatmap - Correlation Analysis
## Objective
A heatmap is used to visualize correlations between numerical variables.
## Why Heatmap?
It helps identify:
- Strong positive correlations
- Strong negative correlations
- Relationships between features
In the Titanic dataset, we analyze relationships among:
- Survived
- Pclass
- Age
- Fare


````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(8,6))

sns.heatmap(
    df.corr(numeric_only=True),
    annot=True,
    cmap="coolwarm"
)

plt.title("Correlation Heatmap")
plt.show()
````

# Count Plot - Gender Distribution
## Objective
Count the number of male and female passengers.
## Why Count Plot?
Count plots show the frequency of categories.
This visualization helps understand gender distribution in the Titanic dataset.

````python
sns.countplot(x="Sex", data=df)

plt.title("Passenger Count by Gender")
plt.show()
````

# Bar Plot - Average Fare by Passenger Class

## Objective
Compare average ticket fare across passenger classes.

## Why Bar Plot?
Bar plots compare numerical values across categories.

Passenger classes:
- First Class
- Second Class
- Third Class

````python
sns.barplot(
    x="Pclass",
    y="Fare",
    data=df
)

plt.title("Average Fare by Passenger Class")
plt.show()
````

# Violin Plot - Fare Distribution by Survival Status
## Objective
Compare fare distributions between survivors and non-survivors.
## Why Violin Plot?
A violin plot combines:
- Box plot
- Density plot
It shows both spread and distribution shape.

````python
sns.violinplot(
    x="Survived",
    y="Fare",
    data=df
)

plt.title("Fare Distribution by Survival Status")
plt.show()
````

# Pair Plot - Multi-Variable Analysis
## Objective
Study relationships among multiple numerical features simultaneously.
## Why Pair Plot?
Pair plots generate:
- Scatter plots
- Histograms
- Correlation insights
Variables analyzed:
- Age
- Fare
- Pclass
- Survived

````python
sns.pairplot(
    df[["Age","Fare","Pclass","Survived"]]
)

plt.show()
````

# Density Plot (KDE Plot) - Age Distribution
## Objective
Visualize the probability density distribution of passenger ages.
## Why Density Plot?
A Density Plot (Kernel Density Estimation - KDE) provides a smooth representation of data distribution.
Benefits:
- Shows the shape of the distribution
- Identifies peaks and clusters
- Easier to interpret than histograms in some cases
In this analysis, we examine how passenger ages are distributed on the Titanic.

````python
import seaborn as sns
import matplotlib.pyplot as plt

sns.kdeplot(
    data=df,
    x="Age",
    fill=True
)

plt.title("Density Plot of Passenger Ages")
plt.xlabel("Age")
plt.ylabel("Density")
plt.show()
````

# Histogram - Passenger Age Distribution
## Objective
Visualize how passenger ages are distributed.
## Why Histogram?
Histograms show:
- Frequency distribution
- Data spread
- Common age ranges

Useful for understanding demographics.

````python
sns.histplot(
    df["Age"],
    bins=20,
    kde=True
)

plt.title("Age Distribution")
plt.show()
````

# Scatter Plot - Age vs Fare
## Objective
Analyze the relationship between passenger age and ticket fare.
## Why Scatter Plot?
Scatter plots reveal:
- Trends
- Clusters
- Outliers
- Relationships between numerical variables

````python
sns.scatterplot(
    x="Age",
    y="Fare",
    data=df
)

plt.title("Age vs Fare")
plt.show()
````
