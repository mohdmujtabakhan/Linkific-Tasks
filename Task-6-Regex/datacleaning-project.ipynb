# Cleaning Messy Customer Data Using Regex

## Objectives
- Remove special characters
- Remove extra spaces
- Compare data before and after cleaning
- Visualize results using Seaborn

## importing necessary libraries 

```python
import pandas as pd
import re
import seaborn as sns
import matplotlib.pyplot as plt
```

## created dataframe of five names to clean name leanths using regex

```python
data = {
    "Name": [
        "John@@ Smith",
        "  Sara### Khan",
        "David!!! Lee",
        "Emma$$ Watson",
        "Mike%%% Brown"
    ]
}

df = pd.DataFrame(data)

df
```

## Calculate Name Length Before Cleaning using str.len() function

```python
df["Before_Length"] = df["Name"].str.len()

df
```

## Before Cleaning Visualization

```python

import seaborn as sns
plt.figure(figsize=(6,4))

sns.histplot(df["Before_Length"], bins=5)

plt.title("Name Length Before Cleaning")

plt.show()
```

## Regex Cleaning
### Pattern 1
`[^A-Za-z ]`
Removes:
- @
- #
- !
- $
- %
### Pattern 2
`\s+`
Removes extra spaces.

```python
import re
df["Name"] = df["Name"].apply(
    lambda x: re.sub(r'[^A-Za-z ]', '', x)
)
```

```python
df["Name"] = df["Name"].apply(
    lambda x: re.sub(r'\s+', ' ', x)
)

df["Name"] = df["Name"].str.strip()

df
```

## Calculate Name Length After Cleaning

```python
df["After_Length"] = df["Name"].str.len()

df
```

## After Cleaning Visualization

```python
plt.figure(figsize=(6,4))

sns.histplot(df["After_Length"], bins=5)

plt.title("Name Length After Cleaning")

plt.show()
```
