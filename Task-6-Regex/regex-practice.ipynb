## What is Regex?

Regular Expression (Regex) is a sequence of characters that defines a search pattern. It is commonly used for:

- Extracting data from text
- Validating input formats
- Searching patterns
- Data cleaning and preprocessing
- Text analysis

## Python Module Used

```python
import re

# Extract Email Addresses using Regex
## Objective
Extract all email addresses present in a given text.
## Regex Pattern Used
`\S+@\S+\.\S+`
### Explanation
- `\S+` → Matches one or more non-space characters.
- `@` → Matches the @ symbol.
- `\.` → Matches the dot (.) character.
- `\S+` → Matches the domain extension.



````python
import re

text = """
Contact us at support@gmail.com
or admin@yahoo.com
"""

emails = re.findall(r'\S+@\S+\.\S+', text)

print("Email Addresses:")
print(emails)
````

# Extract Phone Numbers using Regex
## Objective
Extract 10-digit phone numbers from text.
## Regex Pattern Used
`\d{10}`
### Explanation
- `\d` → Matches any digit (0-9).
- `{10}` → Exactly 10 occurrences.


````python
import re

text = "Call me at 9876543210 or 9123456789"

phones = re.findall(r'\d{10}', text)

print("Phone Numbers:")
print(phones)
````

# Extract Dates in DD/MM/YYYY Format
## Objective
Find dates written in DD/MM/YYYY format.
## Regex Pattern Used
`\d{2}/\d{2}/\d{4}`
### Explanation
- First `\d{2}` → Day
- Second `\d{2}` → Month
- `\d{4}` → Year


````python
import re

text = "Meeting dates are 15/06/2026 and 20/07/2026"

dates = re.findall(r'\d{2}/\d{2}/\d{4}', text)

print("Dates:")
print(dates)
````

# Extract Dates in YYYY-MM-DD Format
## Objective
Find dates written in ISO format.
## Regex Pattern Used
`\d{4}-\d{2}-\d{2}`
### Explanation
- `\d{4}` → Year
- `\d{2}` → Month
- `\d{2}` → Day


````python
import re

text = "Project started on 2026-06-15 and ends on 2026-07-20"

dates = re.findall(r'\d{4}-\d{2}-\d{2}', text)

print("Dates:")
print(dates)
````

# Extract URLs using Regex
## Objective
Extract website links from text.
## Regex Pattern Used
`https?://\S+`
### Explanation
- `https?` → Matches http or https.
- `://` → Matches URL separator.
- `\S+` → Matches remaining URL characters.


````python
import re

text = """
Visit https://google.com
Visit https://github.com
"""

urls = re.findall(r'https?://\S+', text)

print("URLs:")
print(urls)
````

# Extract Hashtags using Regex
## Objective
Extract hashtags from social media text.
## Regex Pattern Used
`#\w+`
### Explanation
- `#` → Matches hashtag symbol.
- `\w+` → Matches letters, digits, and underscores.


````python
import re

text = "#Python #DataScience #MachineLearning"

hashtags = re.findall(r'#\w+', text)

print("Hashtags:")
print(hashtags)
````

# Extract Usernames using Regex
## Objective
Extract usernames tagged with @ symbol.
## Regex Pattern Used
`@\w+`
### Explanation
- `@` → Matches @ symbol.
- `\w+` → Matches username characters.


````python
import re

text = "@mujtaba and @john attended the workshop"

users = re.findall(r'@\w+', text)

print("Usernames:")
print(users)
````

# Extract Numbers using Regex
## Objective
Extract all numeric values from text.
## Regex Pattern Used
`\d+`
### Explanation
- `\d` → Matches a digit.
- `+` → One or more occurrences.


````python
import re

text = "Order 123 contains 45 items costing 678 rupees"

numbers = re.findall(r'\d+', text)

print("Numbers:")
print(numbers)
````

# Remove Special Characters using Regex
## Objective
Clean text by removing symbols and special characters.
## Regex Pattern Used
`[^a-zA-Z0-9 ]`
### Explanation
- `^` → Negation operator.
- `a-zA-Z` → All alphabets.
- `0-9` → Digits.
- Space is preserved.

````python
import re

text = "Hello!!! Welcome@@@ to ###Python$$$"

clean_text = re.sub(r'[^a-zA-Z0-9 ]', '', text)

print("Original Text:")
print(text)

print("\nCleaned Text:")
print(clean_text)
````

# Remove Extra Spaces using Regex
## Objective
Normalize text by replacing multiple spaces with a single space.
## Regex Pattern Used
`\s+`
## Explanation
- `\s` → Matches whitespace.
- `+` → One or more spaces.


````python
import re

text = "Python      is      awesome"

clean_text = re.sub(r'\s+', ' ', text)

print("Original Text:")
print(text)

print("\nCleaned Text:")
print(clean_text)
````

# Validate Email Address
## Objective
Check whether an email address follows a valid format.
## Regex Pattern Used
`^[\w\.-]+@[\w\.-]+\.\w+$`
### Explanation
- `^` → Start of string.
- `[\w\.-]+` → Username.
- `@` → Mandatory symbol.
- `\.` → Dot separator.
- `$` → End of string.


````python
import re

email = "test@gmail.com"

pattern = r'^[\w\.-]+@[\w\.-]+\.\w+$'

if re.match(pattern, email):
    print("Valid Email")
else:
    print("Invalid Email")
````
