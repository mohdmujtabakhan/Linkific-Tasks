# 1. Sentiment Analysis

Sentiment Analysis is used to determine the emotional tone of a piece of text.

The model predicts whether the sentiment expressed in the text is Positive or Negative.

```python
from transformers import pipeline

classifier = pipeline(
    "sentiment-analysis",
    model="distilbert-base-uncased-finetuned-sst-2-english"
)

result = classifier(
    "This movie was absolutely amazing and enjoyable."
)

print(result)
```

### Explanation

The pre-trained sentiment analysis model analyzes the input text and returns:

- Sentiment Label
- Confidence Score

A higher score indicates greater confidence in the prediction.

# 2. Question Answering

Question Answering models extract answers from a given context.

The model reads the context and identifies the most relevant answer to the question.

```python
from transformers import pipeline

qa = pipeline(
    "question-answering",
    model="distilbert-base-cased-distilled-squad"
)

context = """
Python is a programming language created by Guido van Rossum.
"""

question = "Who created Python?"

result = qa(
    question=question,
    context=context
)

print("Question:", question)
print("Answer:", result["answer"])
```

### Explanation

The Question Answering model searches the provided context and extracts the answer that best matches the question.

# 3. Text Generation (GPT-2)

Text Generation models generate new text based on a given prompt.

GPT-2 is a transformer-based language model capable of generating human-like text.

```python
from transformers import pipeline

generator = pipeline(
    "text-generation",
    model="gpt2"
)

result = generator(
    "Artificial Intelligence is",
    max_length=40,
    num_return_sequences=1
)

print(result[0]["generated_text"])
```

### Explanation

The model receives a text prompt and predicts the next most likely words to generate a complete sentence or paragraph.

# 4. Named Entity Recognition (NER)

Named Entity Recognition identifies important entities present in text such as:

- Person
- Organization
- Location
- Date

```python
from transformers import pipeline

ner = pipeline(
    "ner",
    model="dslim/bert-base-NER",
    aggregation_strategy="simple"
)

text = """
Elon Musk is the CEO of Tesla and SpaceX.
He recently visited India and met officials in New Delhi.
"""

result = ner(text)

for entity in result:
    print(
        f"Entity: {entity['word']}, "
        f"Type: {entity['entity_group']}, "
        f"Score: {entity['score']:.4f}"
    )
```

### Explanation

The NER model identifies and classifies entities within text.

Example:

- Elon Musk → Person
- SpaceX → Organization
- United States → Location

# 5. Text Summarization

Text Summarization generates a shorter version of a longer text while preserving the main meaning.

```python
from transformers import pipeline

summarizer = pipeline(
    "summarization",
    model="sshleifer/distilbart-cnn-12-6"
)

text = """
Artificial Intelligence is transforming industries across the world.
Organizations are using AI for automation, data analysis,
customer support, healthcare applications, and predictive analytics.
AI helps businesses improve efficiency and make better decisions.
"""

summary = summarizer(
    text,
    max_length=40,
    min_length=15,
    do_sample=False
)

print(summary[0]["summary_text"])
```

### Explanation

The summarization model reduces long text into a concise summary while retaining important information.
