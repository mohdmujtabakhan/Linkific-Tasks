# Token Cost Calculation

Large Language Model (LLM) APIs charge users based on the number of tokens processed instead of the number of words. Every interaction with the model consumes two types of tokens:

- **Input Tokens:** Tokens present in the user's prompt.
- **Output Tokens:** Tokens generated by the AI in its response.

The total number of tokens processed is calculated as:

**Total Tokens = Input Tokens + Output Tokens**

Most LLM providers charge a fixed price for every one million tokens processed. Therefore, the estimated API cost can be calculated using the following formula:

**Cost = (Total Tokens ÷ 1,000,000) × Price per Million Tokens**

In the following examples, a hypothetical price of **$0.20 per one million tokens** is used to demonstrate the calculation process.

```python
input_tokens = 500
output_tokens = 300
price_per_million = 0.20

total_tokens = input_tokens + output_tokens
cost = (total_tokens / 1_000_000) * price_per_million

print("Input Tokens:", input_tokens)
print("Output Tokens:", output_tokens)
print("Total Tokens:", total_tokens)
print("Estimated Cost: $", round(cost, 6))
```

## Example 2

Suppose the user asks the model to summarize a large research article.

Input Tokens = 2,000

Output Tokens = 800

Total Tokens = 2,800

Cost = (2,800 ÷ 1,000,000) × 0.20

Estimated Cost = **$0.00056**

This example shows that processing larger documents consumes more tokens and therefore increases the API cost.

```python
input_tokens = 2000
output_tokens = 800
price_per_million = 0.20

total_tokens = input_tokens + output_tokens
cost = (total_tokens / 1_000_000) * price_per_million

print("Input Tokens:", input_tokens)
print("Output Tokens:", output_tokens)
print("Total Tokens:", total_tokens)
print("Estimated Cost: $", round(cost, 6))
```
