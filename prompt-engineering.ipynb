# What is Prompt Engineering?

Prompt Engineering is the process of designing clear and effective prompts to obtain accurate, relevant, and high-quality responses from a Large Language Model (LLM).

The quality of an AI's response depends greatly on how the prompt is written. A well-structured prompt provides context, specifies the desired output, and reduces ambiguity, leading to better and more reliable responses.

Prompt Engineering is widely used in AI applications such as chatbots, code generation, document summarization, information extraction, content creation, and decision support systems.

# Zero-shot Prompting

Zero-shot prompting means asking the model to perform a task without providing any examples. The model relies entirely on its pre-trained knowledge to generate a response.

This technique is useful for simple and commonly understood tasks.

```python
from groq import Groq

client = Groq(
    api_key="YOUR_GROQ_API_KEY"
)

response = client.chat.completions.create(
    messages=[
        {
            "role": "user",
            "content": "Explain Machine Learning in simple words."
        }
    ],
    model="llama-3.3-70b-versatile"
)

print(response.choices[0].message.content)
```

# Few-shot Prompting

Few-shot prompting provides one or more examples before asking the model to complete the actual task.

These examples help the model understand the expected output format and improve consistency.

```python
response = client.chat.completions.create(
    messages=[
        {
            "role":"user",
            "content":"""
English: Hello
French: Bonjour

English: Thank You
French: Merci

English: Good Morning
French:
"""
        }
    ],
    model="llama-3.3-70b-versatile"
)

print(response.choices[0].message.content)
```

# Chain-of-Thought Prompting

Chain-of-Thought (CoT) prompting encourages the model to explain its reasoning before providing the final answer.

This technique is especially useful for mathematics, programming, logical reasoning, and complex problem solving.

```python
response = client.chat.completions.create(
    messages=[
        {
            "role":"user",
            "content":"A train travels 60 km per hour for 4 hours. Calculate the total distance and explain your reasoning step by step."
        }
    ],
    model="llama-3.3-70b-versatile"
)

print(response.choices[0].message.content)
```

# Information Extraction Exercise

Information extraction identifies specific details from unstructured text such as names, email addresses, phone numbers, dates, locations, or skills.

This technique is useful for document processing, resume analysis, and data mining.

```python
def ask_llm(prompt):
    response = client.chat.completions.create(
        messages=[
            {
                "role": "user",
                "content": prompt
            }
        ],
        model="llama-3.3-70b-versatile"
    )

    return response.choices[0].message.content

    
print(ask_llm("""
Text:

John Smith works as a Software Engineer at Microsoft.
His email address is johnsmith@gmail.com.
His phone number is 9876543210.

Extract only:
1. Name
2. Email Address
"""))
```

# Comparison of Prompting Techniques

| Prompting Technique | Description | Best Use Case |
|--------------------|-------------|---------------|
| Zero-shot Prompting | Performs a task without examples. | Simple and common tasks |
| Few-shot Prompting | Uses one or more examples before asking the actual question. | Structured outputs and classification |
| Chain-of-Thought Prompting | Encourages the model to explain its reasoning before providing the answer. | Mathematics, coding, and logical reasoning |

Each prompting technique serves a different purpose. Selecting the appropriate technique improves the quality and accuracy of responses generated by Large Language Models.
