# Introduction to NLP

Natural Language Processing (NLP) is a branch of Artificial Intelligence that enables computers to understand, analyze, and process human language.
NLP is used in many real-world applications such as:
- Chatbots
- Language Translation
- Sentiment Analysis
- Spam Detection
- Voice Assistants

Since machine learning algorithms cannot understand text directly, NLP techniques are used to convert text into meaningful numerical representations.

# Import Required Libraries

The following libraries are used for text preprocessing:
- NLTK (Natural Language Toolkit)
- Scikit-Learn

```python
import nltk

from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
```

# Download NLTK Resources

NLTK requires additional resources for tokenization, stopword removal, and lemmatization.

```python
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
```

# Sample Text

A sample sentence is used to demonstrate different NLP preprocessing techniques.

```python
text = "The movie was amazing and I really enjoyed watching it."

print(text)
```

# Tokenization

Tokenization is the process of splitting a sentence into individual words called tokens.

Example:

"The movie was amazing"

↓

["The", "movie", "was", "amazing"]

```python
tokens = word_tokenize(text)

print(tokens)
```

# Stopword Removal

Stopwords are common words that usually do not contribute much meaning to a sentence.

Examples:

- the
- is
- and
- was

Removing stopwords helps reduce noise in the text.

```python
stop_words = set(stopwords.words("english"))

filtered_words = [
    word for word in tokens
    if word.lower() not in stop_words
]

print(filtered_words)
```

# Stemming

Stemming reduces words to their root form.

Examples:

- playing → play
- played → play
- player → play

Stemming helps reduce vocabulary size.

```python
stemmer = PorterStemmer()

stemmed_words = [
    stemmer.stem(word)
    for word in filtered_words
]

print(stemmed_words)
```

# Lemmatization

Lemmatization converts words into their meaningful dictionary form.

It is generally more accurate than stemming.
Examples:
- running → run
- better → good
- studies → study

```python
lemmatizer = WordNetLemmatizer()

lemmatized_words = [
    lemmatizer.lemmatize(word)
    for word in filtered_words
]

print(lemmatized_words)
```

# Bag of Words (BoW)

Bag of Words converts text into numerical vectors based on word frequency.

Machine learning algorithms require numerical data, so text must be converted into numbers.

The frequency of each word becomes a feature.

```python
from sklearn.feature_extraction.text import CountVectorizer

documents = [
    "I love machine learning",
    "Machine learning is powerful"
]

vectorizer = CountVectorizer()

bow = vectorizer.fit_transform(documents)

print(vectorizer.get_feature_names_out())
print(bow.toarray())
```

# TF-IDF (Term Frequency-Inverse Document Frequency)

TF-IDF is an advanced text representation technique.

It assigns higher importance to unique and meaningful words while reducing the importance of commonly occurring words.

TF-IDF often produces better results than Bag of Words because it considers both word frequency and word importance.

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer()

tfidf_matrix = tfidf.fit_transform(documents)

print(tfidf.get_feature_names_out())
print(tfidf_matrix.toarray())
```
